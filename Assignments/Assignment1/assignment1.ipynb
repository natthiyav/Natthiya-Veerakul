### Task 1: A list of dictionaries (1p)

You are given a list of raw transaction records, where each record is a dictionary:

```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
```

Write a script that performs the following steps:

a) Filter out invalid transactions where the price is zero or the quantity is negative.

b) For each remaining transaction, compute a new field `total` = `price * quantity`.

c) Group the valid transactions by `region` and compute the total revenue per region. Store the result in a dictionary.

d) Print the region with the highest total revenue and its value using an f-string.

````python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"}]
#a
valid_transactions = [t for t in transactions
    if t["price"] > 0 and t["quantity"] >= 0]

#b
for t in valid_transactions:
    t["total"] = t["price"] * t["quantity"]

#c
revenue_by_region = {}

for t in valid_transactions:
    region = t["region"]
    
    if region not in revenue_by_region:
        revenue_by_region[region] = 0
    
    revenue_by_region[region] += t["total"]

#d
max_region = max(revenue_by_region, key=revenue_by_region.get)
max_value = revenue_by_region[max_region]

print(f"The region with the highest revenue is {max_region} with total revenue {max_value}.")
````

### Task 2: Creating a function (1p)

You are given the following list of daily sales figures (in euros) for a two-week period:

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]
```

Write a function called `summarise_sales` that takes a list of daily sales values and returns a dictionary containing:

`"total"`: the sum of all daily sales

`"average"`: the mean daily sales value, rounded to two decimals

`"best_day"`: the index (starting from 1) of the day with the highest sales

`"worst_day"`: the index (starting from 1) of the day with the lowest sales

`"above_average_count"`: the number of days where sales exceeded the average

Call the function with the list above and print the returned dictionary.

````python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

def summarise_sales(daily_sales):
    # Total sales
    total = sum(daily_sales)
    
    # Average sales
    average = round(total / len(daily_sales), 2)
    
    # Best day (highest sales)
    best_day = daily_sales.index(max(daily_sales)) + 1  # +1 to start counting from day 1
    
    # Worst day (lowest sales)
    worst_day = daily_sales.index(min(daily_sales)) + 1
    
    # Number of days above average
    above_average_count = sum(1 for sale in daily_sales if sale > average)
    
    # Return the dictionary
    return {
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count
    }

# Call the function and print result
result = summarise_sales(sales)
print(result)
````

### Task 4: Class and methods (1p)

Design a class called `Portfolio` that represents an investment portfolio. The constructor is provided for you:

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}
```

Each key in `holdings` is a ticker symbol (string) and each value is a dictionary with keys `"shares"` and `"price_per_share"`. Add the following methods to the class:

a) A method `add_asset(ticker, shares, price_per_share)` that adds a new asset or updates the number of shares if the ticker already exists (in which case the price should also be updated).

b) A method `total_value()` that returns the total market value of the portfolio.

c) A method `most_valuable()` that returns the ticker of the asset with the highest total value (shares × price).

d) A `__str__` method that returns a readable summary of the portfolio, including the owner's name, the number of assets, and the total value.

Create an instance, add at least three assets, and demonstrate all methods with printed output.

````python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}

#a
    def add_asset(self, ticker, shares, price_per_share):
        if ticker in self.holdings:
            self.holdings[ticker]["shares"] += shares
            self.holdings[ticker]["price_per_share"] = price_per_share
        else:
            self.holdings[ticker] = {
                "shares": shares,
                "price_per_share": price_per_share
            }

#b
    def total_value(self):
        total = 0
        for asset in self.holdings.values():
            total += asset["shares"] * asset["price_per_share"]
        return total

#c
    def most_valuable(self):
        highest_ticker = None
        highest_value = 0

        for ticker, asset in self.holdings.items():
            value = asset["shares"] * asset["price_per_share"]
            if value > highest_value:
                highest_value = value
                highest_ticker = ticker

        return highest_ticker

#d
    def __str__(self):
        return (f"Portfolio owner: {self.owner}\n"
                f"Number of assets: {len(self.holdings)}\n"
                f"Total value: {self.total_value():.2f}")
````

````python
#Example
my_portfolio = Portfolio("Natthiya")

# 2. Add at least three assets
my_portfolio.add_asset("AAPL", 10, 150.00)
my_portfolio.add_asset("MSFT", 5, 310.50)
my_portfolio.add_asset("GOOGL", 20, 135.25)

print("--- Initial Portfolio ---")
print(my_portfolio) # Demonstrates __str__
print(f"Total Market Value: ${my_portfolio.total_value():.2f}") # Demonstrates total_value()
print(f"Most Valuable Ticker: {my_portfolio.most_valuable()}") # Demonstrates most_valuable()

print("\n--- After buying 5 more AAPL shares at $155.00 ---")
my_portfolio.add_asset("AAPL", 5, 155.00)

print(my_portfolio)
print(f"New AAPL Holdings: {my_portfolio.holdings['AAPL']}")
````

### Task 5: Importing a module (1p)

You are given the following list of order dates (as strings) and product names:

```python
orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]
```

Using the `datetime` module, parse each date string into a `datetime.date` object. Compute and print the number of days between the earliest and latest order.

````python
from datetime import datetime

orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]

# Parse dates into datetime.date objects
dates = [datetime.strptime(order[0], "%Y-%m-%d").date() for order in orders]

# Find earliest and latest dates
earliest = min(dates)
latest = max(dates)

# Compute difference
difference = (latest - earliest).days

print("Earliest date:", earliest)
print("Latest date:", latest)
print("Number of days between:", difference)
````

### Task 8: More Pandas (1p)

A company records daily website traffic over four weeks. Generate the data by running the following script first:

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)
```

Using this generated data, perform the following:

a) Create a DataFrame with columns: `date`, `visitors`, `conversion_rate`, and `avg_order_value`. Set `date` as the index.

b) Compute two new columns: `conversions` (= visitors × conversion_rate, rounded to the nearest integer) and `daily_revenue` (= conversions × avg_order_value).

````python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)

#a
df = pd.DataFrame({
    "date": days,
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value})

df = df.set_index("date")
print("DataFrame:\n", df.head(), "\n")
````

````python
#b
df["conversions"] = (df["visitors"] * df["conversion_rate"]).round().astype(int)
df["daily_revenue"] = df["conversions"] * df["avg_order_value"]

print("DataFrame with new columns:\n", df.head())
````
